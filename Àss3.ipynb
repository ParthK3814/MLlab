```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification, make_regression
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
from sklearn.metrics import accuracy_score, mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
X_cls, y_cls = make_classification(n_samples=200, n_features=2, n_informative=2,
                                   n_redundant=0, n_classes=2, random_state=42)
Xc_train, Xc_test, yc_train, yc_test = train_test_split(X_cls, y_cls, test_size=0.3, random_state=42)

```

```python

ks = [1, 3, 5, 7, 9]
cls_accuracies = []
for k in ks:
    knn_cls = KNeighborsClassifier(n_neighbors=k)
    knn_cls.fit(Xc_train, yc_train)
    y_pred = knn_cls.predict(Xc_test)
    cls_accuracies.append(accuracy_score(yc_test, y_pred))
```

```python

plt.plot(ks, cls_accuracies, marker='o')
plt.title("KNN Classification: Accuracy vs K")
plt.xlabel("K")
plt.ylabel("Accuracy")
plt.show()
```

```python

X_reg, y_reg = make_regression(n_samples=200, n_features=1, noise=20, random_state=42)
Xr_train, Xr_test, yr_train, yr_test = train_test_split(X_reg, y_reg, test_size=0.3, random_state=42)

```

```python

reg_mse = []
for k in ks:
    knn_reg = KNeighborsRegressor(n_neighbors=k)
    knn_reg.fit(Xr_train, yr_train)
    y_pred = knn_reg.predict(Xr_test)
    reg_mse.append(mean_squared_error(yr_test, y_pred))
```

```python

plt.plot(ks, reg_mse, marker='o')
plt.title("KNN Regression: MSE vs K")
plt.xlabel("K")
plt.ylabel("MSE")
plt.show()
```
